```python
import threading

def print_even():
    for i in range(2, 51, 2):
        print(i)

def print_odd():
    for i in range(1, 51, 2):
        print(i)

t1 = threading.Thread(target=print_even)
t2 = threading.Thread(target=print_odd)


t1.start()
t2.start()

t1.join()
t2.join()
```

```python
import threading


def print_digits():
    for i in range(1, 27):
        print(i)


def print_lowercase():
    for i in range(97, 123):
        print(chr(i))


def print_uppercase():
    for i in range(65, 91):
        print(chr(i))


t1 = threading.Thread(target=print_digits)
t2 = threading.Thread(target=print_lowercase)
t3 = threading.Thread(target=print_uppercase)

t1.start()
t2.start()
t3.start()


t1.join()
t2.join()
t3.join()
```

```python
import threading


def print_squares():
    for i in range(1, 21):
        print(i ** 2)

def print_cubes():
    for i in range(1, 21):
        print(i ** 3)

t1 = threading.Thread(target=print_squares)
t2 = threading.Thread(target=print_cubes)


t1.start()
t2.start()


t1.join()
t2.join()
```
